### Importing necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Generating a sinusoid
Here I'm generating a sin-wave with amplitude 10

```python
t = np.linspace(-5, 5, 1000)
sin = 10 * np.sin(2 * np.pi * 2 * t)

plt.plot(t, sin)
plt.title('Sinusoidal Wave')
plt.ylabel('Wavefunction')
plt.xlabel('Time')
plt.show()
```

### Generating a ramp wave
Here I'm generating a ramp wave that starts moving upwards at t=0 with slope=2

```python
x = np.hstack([np.linspace(0, 5, 333), np.linspace(0, 5, 333), np.linspace(0, 5, 334)])

plt.plot(t, x)
plt.title('Ramp Wave')
plt.ylabel('Wavefunction')
plt.xlabel('Time')
plt.show()
```

### Mixing the two signals
This cell mixes the two signals (sinusoid and ramp) according to the given mixing matrix)

```python
productMatrix = np.array([[0.5, 1], [1, 0.5]])@np.vstack([sin, x])
productMatrix = productMatrix-productMatrix.mean(axis=1, keepdims=True)
```

### Retrieving original signals from the mixed waveform
Here is a cell containing helper functions containing the standard ICA algorithm, along with a cell to run the functions to predict original waveforms

```python
def whiten(X):
    eigvals, eigvects = np.linalg.eigh(np.cov(X))
    return np.dot(eigvects, np.dot(np.sqrt(np.linalg.inv(np.diag(eigvals))), np.dot(eigvects.T, X)))

def getNewVects(vects, X):
    a = (X * np.tanh(np.dot(vects.T, X))).mean(axis=1)
    b = (1-np.tanh(np.dot(vects.T, X))*np.tanh(np.dot(vects.T, X))).mean()*vects
    return (a - b)/np.sqrt(((a - b) ** 2).sum())

def independentComponentAnalysis(signal, maxiter, minDist):
    signal = signal - signal.mean(axis=1, keepdims=True)
    signal = whiten(signal)
    vects_final = np.zeros((signal.shape[0], signal.shape[0]), dtype=signal.dtype)
    for i in range(signal.shape[0]):
        vects = np.random.rand(signal.shape[0])
        for j in range(maxiter):
            vects_new = getNewVects(vects, signal)
            if (i > 0):
                vects_new = vects_new - np.dot(np.dot(vects_new, vects_final[:i].T), vects_final[:i])
            dist = np.abs(1-np.abs((vects_new * vects).sum()))
            vects = vects_new
            if minDist > dist:
                break 
        vects_final[i] = vects
    return np.dot(vects_final, signal)
```

```python
independentSignals = independentComponentAnalysis(productMatrix, 100, 0.00001)
fig = plt.figure()
for i, signal in enumerate(independentSignals):
    plt.plot(signal)
    plt.title(f"Predicted component signal ({i+1})")
    plt.show()
```

(Comments: the separated waveforms are remarkably similar to the original signals)
